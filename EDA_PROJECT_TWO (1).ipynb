which hosts are the buiest and why is the reason?

```python
busy_hosts = df_airbnb.groupby(['host_id','host_name','room_type'])['number_of_reviews'].max().reset_index()
busy_hosts = busy_hosts.sort_values(by= 'number_of_reviews',ascending = False).head(20)
busy_hosts
```

```python
name_hosts = busy_hosts['host_name']
review_got = busy_hosts['number_of_reviews']
fig = plt.figure(figsize = (20,8))
plt.bar(name_hosts,review_got,color = 'red', width = 0.6)
plt.xlabel('name of the host ')
plt.ylabel('Review')
plt.title('Busiest Host in term of reviews')
plt.show()
```

is there any traffic difference among different property and what could be the reason for it?

```python
df_airbnb.head(10)
```

```python
traffic_reviews = df_airbnb.groupby(['property_type','room_type'])['minimum_nights'].count().reset_index()
traffic_reviews = traffic_reviews.sort_values(by = 'minimum_nights',ascending=False ).head(60)
traffic_reviews
```

```python
Traffic_reviews = traffic_reviews['room_type']
room_stayed = traffic_reviews['minimum_nights']
fig = plt.figure(figsize =(15,8))
plt.bar(Traffic_reviews,room_stayed,color = 'green',width = 0.3)
plt.xlabel('Room Type ')
plt.ylabel('Minimum Nights ')
plt.title('Traffic proprty based on Minimum Nights booked')
plt.show()
```

From this visualization we found that most of the people likly to stay at Entire home and Private room which are present in rental unit , private room in home , condo,guesthouse and also vistor referring stay in room which listing price is less
